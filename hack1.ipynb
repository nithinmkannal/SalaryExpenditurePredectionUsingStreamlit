```python
import pandas as pd

# Assuming data_df is a Pandas DataFrame

for column in data_df.columns:
    # Fill NA values with the average of the previous and next non-NA values
    data_df[column].fillna(data_df[column].interpolate(method='linear'), inplace=True)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np



# Load data into a Pandas DataFrame (assuming data_df is already loaded)

# Iterate through predictor columns
for predictor_column in list(data_df.columns):
    if predictor_column != "STATES":
        # Prepare training data
        X_train = data_df[["STATES"]].iloc[:7]  # Feature variable
        y_train = data_df[predictor_column].iloc[:7]  # Target variable

        # Create linear regression model
        model = LinearRegression()
        model.fit(X_train, y_train)

        # Define X values for prediction
        X_to_predict = [[2015], [2016]]

        # Make predictions
        predictions = model.predict(X_to_predict)

        # Display predictions
        print(f"Predictions of SALARY EXPENDITURE for {predictor_column} when  year = 2014-2015, 2015-2016: {predictions}")
        # Display linear regression equation
        slope = model.coef_[0]
        intercept = model.intercept_

        equation = f"y = {slope:.2f} * x + {intercept:.2f}"
        print(f"Linear Regression Equation for {predictor_column}: {equation}")

       # Display predictions
        print(f"Predictions for {predictor_column} when year = 2015, 2016: {predictions}")
        x = np.linspace(2013, 2020, 100)
        y = slope * x + intercept
        plt.figure(figsize=(8, 6))
        plt.plot(x, y, color="red")

        # Set labels and title
        plt.xlabel("years")
        plt.ylabel("salary expenditure over years")
        plt.title(f"Graph of salry expenditure for {predictor_column}")
        plt.grid(True)
        plt.show()





```
